## **PREGUNTA 1** <font color="red">Daniel</font>
¿La distribución de cnt puede tratarse como aproximadamente normal, realiza el histrograma y el gráfico Q-Q?

# **Respuesta pregunta 1**

**Histograma:** normalmente muestra una distribución sesgada a la derecha (asimetría positiva). Es decir, no se ve como una campana de Gauss sino más concentrada en valores bajos-medios con una cola larga hacia valores altos.

**Gráfico Q-Q:** los puntos suelen desviarse bastante de la línea recta, especialmente en las colas, lo que indica que la distribución no es normal.

Pruebas estadísticas (Shapiro-Wilk o Kolmogorov-Smirnov): en general rechazan la hipótesis de normalidad (p < 0.05).

**En conclusión:** la variable **cnt** no puede considerarse aproximadamente normal, aunque en análisis de datos es común aplicar transformaciones (logarítmica o Box-Cox) si se quiere aproximar a una distribución normal.

```python
# 1. Visualización
plt.figure(figsize=(14, 5))
sns.histplot(df['cnt'], kde=True)
plt.title("2. Distribución de cnt")
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Histograma de la variable 'cnt'
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(df["cnt"], bins=40, edgecolor="black", alpha=0.7)
plt.title("Histograma de cnt")
plt.xlabel("cnt")
plt.ylabel("Frecuencia")

# Gráfico Q-Q
plt.subplot(1,2,2)
stats.probplot(df["cnt"], dist="norm", plot=plt)
plt.title("Q-Q plot de cnt")

plt.tight_layout()
plt.show()
```

### **Interpretación pregunta 1**

La variable **`cnt`** no presenta una distribución aproximadamente normal, ya que en el histograma se observa un claro sesgo positivo con mayor concentración de valores en niveles bajos y medios, mientras que los valores altos son menos frecuentes, generando una cola alargada hacia la derecha. Esta tendencia se confirma en el gráfico Q-Q, donde los puntos se alejan de la diagonal especialmente en los extremos, evidenciando que los datos no siguen el patrón de una distribución normal. Por lo tanto, se concluye que `cnt` no cumple con el supuesto de normalidad, lo que sugiere la necesidad de aplicar transformaciones o emplear métodos estadísticos no paramétricos en su análisis.


## **PREGUNTA 2** <font color="red">Jonathan</font>
¿Cómo co-varía cnt con temp, hum y windspeed? ¿La relación es lineal o solo monótona?. Ayuda: Calcular la correlación de Pearson y de Spearman.

# **Respuesta pregunta 2**

La cantidad de bicicletas alquiladas (cnt) aumenta cuando sube la temperatura, mostrando una relación bastante lineal: entre más calor, más bicicletas se usan. En cambio, con la humedad ocurre lo contrario: a mayor humedad, disminuyen los alquileres, aunque esta relación no es tan perfectamente lineal, sino más bien una tendencia general (monótona).

Finalmente, con la velocidad del viento, prácticamente no se observa una relación fuerte; los cambios en el viento no modifican de manera clara el uso de las bicicletas.

**En conclusión**: la temperatura tiene una relación lineal positiva, la humedad una relación monótona negativa, y el viento prácticamente no influye.

```python
# Selección de columnas
columnas_relacion = ['cnt', 'temp', 'hum', 'windspeed']
df_corr = df[columnas_relacion]

# Cálculo de correlaciones Pearson y Spearman
corr_pearson = df_corr.corr(method='pearson')
corr_spearman = df_corr.corr(method='spearman')

print("Correlación de Pearson:\n", corr_pearson, "\n")
print("Correlación de Spearman:\n", corr_spearman, "\n")

# Visualización con mapas de calor
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(corr_pearson, annot=True, cmap="coolwarm", center=0, ax=axes[0])
axes[0].set_title("Correlación de Pearson (lineal)")

sns.heatmap(corr_spearman, annot=True, cmap="coolwarm", center=0, ax=axes[1])
axes[1].set_title("Correlación de Spearman (monótona)")

plt.show()
```

### **Interpretación pregunta 2**

La cantidad de bicicletas alquiladas (cnt) aumenta claramente con la temperatura, disminuye con la humedad y apenas se ve afectada por la velocidad del viento. La relación con la temperatura es más lineal, mientras que con la humedad es solo una tendencia general (monótona).

## **PREGUNTA 4** <font color="red">Mercy</font>
En días laborales, ¿la media de cnt ≥ 4.500? (plantea y resuelve un test de hipótesis de una muestra).

# **Hipotesis**

Queremos verificar si, en días laborales, el número medio de alquileres de bicicletas (cnt) es mayor o igual que 4.500.

Hipótesis nula (H₀): μ = 4500

Hipótesis alternativa (H₁): μ > 4500

Se trata de una prueba t de una muestra unilateral (cola derecha).

```python
from scipy import stats
import numpy as np

# Datos de cnt en días laborales
cnt_laborales = df[df['workingday'] == 1]['cnt']

# Valor de referencia
valor_hipotesis = 4500

# Estadístico t y p-valor (unilateral, cola derecha)
t_stat, p_val = stats.ttest_1samp(cnt_laborales, valor_hipotesis)
p_val_unilateral = p_val / 2 if t_stat > 0 else 1 - (p_val / 2)

print("Estadístico t:", t_stat)
print("p-valor unilateral:", p_val_unilateral)
print("Media muestral:", np.mean(cnt_laborales))
```

### **Interpretación pregunta 4**

La prueba de hipótesis buscaba determinar si, en días laborales, el número promedio de alquileres de bicicletas (cnt) es mayor o igual a 4.500. Al aplicar el test t de una muestra, se obtuvo que la media muestral de cnt en días laborales está claramente por encima de ese valor de referencia. Además, el p-valor unilateral resultó muy pequeño (menor a 0.05), lo cual indica evidencia estadísticamente significativa para rechazar la hipótesis nula.

Los datos muestran que en días laborales el promedio de bicicletas alquiladas supera los 4.500, lo que refleja una alta demanda en la jornada laboral.

## **PREGUNTA 5** <font color="red">Fabian</font>
¿Cuál es un IC bootstrap para la mediana de cnt y cómo cambia tu mensaje si la distribución es sesgada?

# **Respuesta pregunta 5**

Un IC bootstrap para la mediana de cnt es un rango de valores que estima dónde está la verdadera mediana usando muchas simulaciones con los mismos datos.

- Si la distribución es simétrica, el intervalo queda bastante centrado y se interpreta con confianza.
- Si la distribución es sesgada, el intervalo se inclina hacia un lado y el mensaje cambia: el bootstrap permite reflejar mejor esa asimetría y dar una estimación más realista de la mediana.

```python
import numpy as np

# Datos
x = df['cnt'].values

# Bootstrap
n_iter = 5000
rng = np.random.default_rng(123)
medianas = [np.median(rng.choice(x, size=len(x), replace=True)) for _ in range(n_iter)]

# Intervalo de confianza
ic_low, ic_high = np.percentile(medianas, [2.5, 97.5])

print("Mediana muestral:", np.median(x))
print("IC 95% bootstrap para la mediana:", (ic_low, ic_high))
```

# Interpretación pregunta 5

El intervalo de confianza bootstrap muestra el rango donde probablemente se encuentra la mediana real de los alquileres diarios de bicicletas. Si la distribución está sesgada, este intervalo no es simétrico y refleja esa asimetría, lo que indica que la mediana es una medida más confiable que el promedio para describir los datos.
